```python
import pandas as pd
import numpy as np
from sklearn.model_selection import KFold,GridSearchCV
from sklearn.neural_network import MLPRegressor
from sklearn.metrics import mean_absolute_error
```

```python
X_train = pd.read_csv(r"C:\Users\dbda\Desktop\project\DataSet\EQ_DATA\2.25\X_train_final_2.25.csv")
y_train = pd.read_csv(r"C:\Users\dbda\Desktop\project\DataSet\EQ_DATA\2.25\y_train_final_2.25.csv")
X_test = pd.read_csv(r"C:\Users\dbda\Desktop\project\DataSet\X_test_final.csv")
y_train = y_train['time_to_failure']
```

```python
mlp = MLPRegressor(random_state=2020,early_stopping=True,n_iter_no_change=8,tol=0.01)
kfold = KFold(n_splits=5,random_state=2020)

lr = np.linspace(0.001,0.010,10)
hidden_layer_sizes=[(10,20,40,50,5),(10,10,5,20,20),(10,12,14,20,20,6,8,12,8)]
batch = [10,20,30,50,150,300,350]

params = dict(hidden_layer_sizes=hidden_layer_sizes,learning_rate_init=lr,batch_size=batch)

cv = GridSearchCV(mlp,cv=kfold,scoring='neg_mean_absolute_error',verbose=3,param_grid=params,)
```

```python
cv.fit(X_train,y_train)
```

```python
print(cv.best_score_)
print(cv.best_params_)
```

```python
best = cv.best_estimator_
y = best.predict(X_test)
```

```python
y = pd.DataFrame(y)
y.to_csv(r"C:\Users\dbda\Desktop\project\MLP1_2.25.csv")
```
